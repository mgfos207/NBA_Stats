```python
df_rank = pd.DataFrame()
```

```python
seasons_lists = list(df_joined['Season'].unique())
```

```python
def _get_player_rank(df, cols, season):
    agg_map = {col: 'mean' for col in cols}
    cols_name = {col: f'{col}_rank' for col in cols}
    return df.loc[(df['Season'] == season),['player_id', 'Season', 'Player', 'MVP'] + cols].groupby(by=['player_id', 'Season', 'Player', 'MVP']).agg(agg_map).rank(pct=True).reset_index().rename(columns=cols_name)
```

```python
#ORB	DRB MP STL	BLK
for season in seasons_lists:
    df_rank = pd.concat([df_rank, _get_player_rank(df_joined, ['PTS', 'AST', 'ORB', 'DRB', 'MP', 'STL', 'BLK'], season)])

df_rank.head()
```

```python
df_rank.loc[df_rank['MVP'] == True].head(20)
```

```python
df_rank['avg_stat_rank'] = (df_rank['PTS_rank'] + df_rank['AST_rank'] + df_rank['ORB_rank'] + df_rank['DRB_rank'] + df_rank['MP_rank'] + df_rank['STL_rank'] + df_rank['BLK_rank']) / 7
```

```python
df_rank.loc[df_rank['MVP'] == True, ['Player', 'avg_stat_rank']].head(20).sort_values(by='avg_stat_rank', ascending=False)
```

```python
df_rank.loc[df_rank['MVP'] == True, ['Player', 'avg_stat_rank']].head(20).sort_values(by='avg_stat_rank', ascending=False).hist()
```
